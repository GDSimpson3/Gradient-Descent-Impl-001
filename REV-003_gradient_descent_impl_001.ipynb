# Load Raw Data

This is a simple 2 Column salary dataset (from kaggle) that looks @ the relationship between years of experience and Salary to draw a regression line that could be used to predict VIA interpolation and extrapolation

```python
import pandas as pd
'''Using Pandas as our datahandling library'''

'''Load in the CSV
 DF Means Dataframe

'''

df = pd.read_csv("https://raw.githubusercontent.com/GDSimpson3/ML-datasets/refs/heads/main/Salary_dataset.csv")

print(df)
```

# Visualise Data



## Scatter Plot of Raw Data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['YearsExperience'], df['Salary'])
plt.title('Years of Experience vs. Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

## Custom Regression Line

```python
import numpy as np

theta_0 = 1.0  # The y-intercept
theta_1 = 1.2  # The slope

X_line = np.linspace(df['YearsExperience'].min(), df['YearsExperience'].max(), 100) # Creates 100 evenly spaced points

# 4. Calculate the predicted Y-values (the regression line)
Y_pred = theta_0 + theta_1 * X_line

plt.plot(X_line, Y_pred, color='red', linewidth=2, label=f'Regression Line: y = {theta_1}x + {theta_0}')

# Add titles and labels
plt.title('Custom Linear Regression Line on Scatter Plot')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.legend()
plt.grid(True)
plt.show()
```

# Mean Squared Error Calculator

we're using the same data throughout, hence no need of DYNAMIC dynamic functions

```python
def MSE(m,b):
  YStream = df['Salary']
  XStream = df['YearsExperience']
  Summation = 0
  for N in range(len(YStream)):

    ObservedValue = YStream[N]
    PredictedValue = (m * XStream[N]) + b

    Error = ObservedValue - PredictedValue
    ErrorSquared = Error ** 2

    Summation += ErrorSquared
    # print(f'Observed Value = {ObservedValue} --- Predicted Value = {PredictedValue} --- Error = {Error}')

  return Summation / len(YStream)
```

# Partial Derivatives
This is to look at the rate of change of the BiVariate MSE with respects to only 1 Parameter

```python
def PartialDerivativeB(m,b):
  YStream = df['Salary']
  XStream = df['YearsExperience']
  Summation = 0
  for N in range(len(YStream)):

    ObservedValue = YStream[N]
    PredictedValue = (m * XStream[N]) + b

    Error = PredictedValue - ObservedValue

    Summation = Summation + Error
    # print(f'Observed Value = {ObservedValue} --- Predicted Value = {PredictedValue} --- Error = {Error}')

  return 2 * Summation / len(YStream)
```

```python
def PartialDerivativeM(m,b):
  YStream = df['Salary']
  XStream = df['YearsExperience']
  Summation = 0
  for N in range(len(YStream)):

    ObservedValue = YStream[N]
    PredictedValue = (m * XStream[N]) + b

    Error = PredictedValue - ObservedValue

    ErrorX = Error * XStream[N]

    Summation = Summation + Error
    # print(f'Observed Value = {ObservedValue} --- Predicted Value = {PredictedValue} --- Error = {Error}')

  return 2 * Summation / len(YStream)
```

# Gradient Descent Iteration Steps

```python
Alpha = 0.01
M = 1 # Any starting Value
B = 1

Plot(M,B)

MaxIterations = 10000

for I in range(MaxIterations):
  M = M - (Alpha * PartialDerivativeM(M,B))
  B = B - (Alpha * PartialDerivativeB(M,B))

print(f'Final Parameters give an MSE Of {MSE(M,B)}, Optimal Parameters are M: {M} and B: {B}')

Plot(M,B)
```
